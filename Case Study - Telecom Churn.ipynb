### 1. Filter high-value customers

```python
high_value_cstmr = Tel_churn.total_rech_amt_goodphase.quantile(0.7)
high_value_cstmr
```

```python
def high_k(x):
    if x < 368.5:
        return 0
    else:
        return 1
    
Tel_churn['high_value_cstmr'] = Tel_churn['total_rech_amt_goodphase'].apply(high_k)
```

```python
Tel_churn.high_value_cstmr.value_counts()
```

### 2. Tag churners and remove attributes of the churn phase

```python
churned = (Tel_churn.total_ic_mou_9 == 0) & (Tel_churn.total_og_mou_9 == 0) & (Tel_churn.vol_2g_mb_9 ==0) & (Tel_churn.vol_3g_mb_9 ==0)
Tel_churn.loc[churned,'churner']=1
```

## Modelling

```python
#Test train split
from sklearn.model_selection import train_test_split
```

```python
# Putting feature variable to X
X = Tel_churn.drop(['churner','mobile_number'], axis=1)
X.head()
```

```python
# Putting response variable to y
y = Tel_churn['churner']
y.head()
```

```python
# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train[] = scaler.fit_transform(X_train[])

X_train.head()
```

```python
### Checking the Churn Rate
churn = (sum(Tel_churn['Churn'])/len(Tel_churn['Churn'].index))*100
churn
```

```python
# Let's see the correlation matrix 
plt.figure(figsize = (20,10))        # Size of the figure
sns.heatmap(Tel_churn.corr(),annot = True)
plt.show()
```

```python
#Running your first model
import statsmodels.api as sm
```

```python
# Logistic regression model
logm1 = sm.GLM(y_train,(sm.add_constant(X_train)), family = sm.families.Binomial())
logm1.fit().summary()
```

```python
#Feature Selection using RFE
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
 # running RFE with 13 variables as output
from sklearn.feature_selection import RFE
rfe = RFE(logreg, 15)            
rfe = rfe.fit(X_train, y_train)
```

```python
rfe.support_
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_sm = sm.add_constant(X_train[col])
logm2 = sm.GLM(y_train,X_train_sm, family = sm.families.Binomial())
res = logm2.fit()
res.summary()
```
